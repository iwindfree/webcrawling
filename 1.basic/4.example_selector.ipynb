지금까지 배운 것을 기본으로 네이버 쇼핑에서 top 100 페이지의 상품들을 읽어 오는 크롤링 소스를 만들어 보겠습니다. 크롬의 개발자 도구에서 요소를 선택하고, 'Copy selector' 기능을 이용하면 좀 더 쉽게 selector 검색 조건을 확인할 수 있다. 물론, 해당 선택한 요소에 대한 selector 이기 때문에 약간의 변형은 필요한 경우가 많다.

```python
import requests
from bs4 import BeautifulSoup
res = requests.get("https://search.shopping.naver.com/best100v2/main.nhn")
soup = BeautifulSoup(res.content,'html.parser')
mydata = soup.select('#cateProductListArea50000000 > ul > li > p')
for item in mydata:
    print(item.get_text())
```
